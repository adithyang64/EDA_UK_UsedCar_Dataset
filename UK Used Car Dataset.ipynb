# Working with Unclean Datasets

```python
df_uc_cclass=pd.read_csv('UnusedCarDataset/unclean cclass.csv')
df_uc_focus=pd.read_csv('UnusedCarDataset/unclean focus.csv')
```

```python
## Concatenating Unclean Datasets
df_uc=pd.concat([
df_uc_cclass,
df_uc_focus
], axis=0)
```

```python
## Cleaning price column
df_uc['price'] = [str(i).replace(",", "").strip() for i in df_uc['price']]
df_uc['price'] = [float(str(i).replace("£", "").strip()) for i in df_uc['price']]
```

```python
# Adding Mpg column as it is missing
df_uc['mpg']= np.nan
#dropping Reference Column as itisnt needed
df_uc.drop('reference',axis=1,inplace=True)
```

```python
df_uc['engine size'].unique()
```

```python
## Renamaing Tax column to Engine Size Column and Data Cleaning 
df_uc.rename(columns={'engine size': 'tax'}, inplace=True)
df_uc['tax'] = [str(i).replace("£", "").strip() for i in df_uc['tax']]
df_uc['tax'] = [str(i).lower().replace("cc", "").strip() for i in df_uc['tax']]
df_uc['tax'] = [str(i).replace(",", "").strip() for i in df_uc['tax']]
df_uc['tax'] = [float(str(i).strip()) for i in df_uc['tax']]
df_uc['tax']
```

```python
# Renaming Engine Size 2 as Engine Size as it conatins the actual Data
df_uc.rename(columns={'engine size2' : 'engineSize'}, inplace= True)
```

```python
df_uc['engineSize'].unique()
```

```python
df_uc['engineSize'] = [str(i).replace('cc','').strip()  for i in df_uc['engineSize']]
df_uc['engineSize'] = [str(i).replace('T','').strip()  for i in df_uc['engineSize']]
df_uc['engineSize'] = [str(i).replace('Unknown','').strip()   for i in df_uc['engineSize']]
df_uc['engineSize'] = pd.to_numeric(df_uc['engineSize'], errors='coerce')

# converting cc to liters -> 1000cc = 1L
df_uc['engineSize'] = df_uc.apply(lambda x: x['engineSize']/1000 if x['engineSize']> 20 else x['engineSize'] , axis=1)
```

```python
df_uc['engineSize'].unique()
```

```python
## Data cleaning of Mileage Columns 
mileage=['mileage','mileage2']
for i in mileage:
    df_uc[i] = [str(i).split('.')[0] for i in df_uc[i]]
    df_uc[i] = [str(i).replace(',', '') for i in df_uc[i]]
    df_uc[i] = [str(i).replace('Unknown', '') for i in df_uc[i]]
    df_uc[i]=df_uc[i].replace('nan', '')
    df_uc[i] = pd.to_numeric(df_uc[i])
```

```python
## Taking the suitable value among the mileage columns
df_uc["mileage"] = df_uc.apply(lambda x: x["mileage"] if x["mileage"]!= np.NaN and x["mileage"]>x["mileage2"] else x["mileage2"], axis=1)
```

```python
df_uc.drop(["mileage2"], inplace=True, axis=1)
```

```python
df_uc.head()
```

```python
df_uc.head()
```

```python
## Tax Columns contains data of both Tax and Engine size - Matching them accordingly to respective columns
df_uc["engineSize"] = df_uc.apply(lambda x: x["tax"] if x["tax"]<4 else x["engineSize"], axis=1)
```

```python
## Marking NaN in Tax Column for Engine Size data which is not required in this column
df_uc["tax"] = df_uc.apply(lambda x: np.NaN if x["tax"]<4 else x["tax"], axis=1)
```

```python
#Combining the  fuel type and fuel type2  as strings and the then remove appending letters
df_uc['fuel type']=df_uc['fuel type'].fillna("NaN")
df_uc['fuel type2']=df_uc['fuel type2'].fillna("NaN")

df_uc[['fuel type','fuel type2']]=df_uc[['fuel type','fuel type2']].astype('str')
df_uc["fuelType"] = df_uc[['fuel type','fuel type2']].agg(', '.join, axis=1)
df_uc["fuelType"] = df_uc["fuelType"].str.replace('NAN, ', '')
df_uc["fuelType"] = df_uc["fuelType"].str.replace(', NAN', '')
df_uc["fuelType"] = df_uc["fuelType"].str.replace(' ', '')
```

```python
def remove_outlier_IQR(df,ft):
    Q1=df[ft].quantile(0.25)
    Q3=df[ft].quantile(0.75)
    IQR=Q3-Q1
    ls=df.index[((df[ft]<(Q1-1.5*IQR)) | (df[ft]>(Q3+1.5*IQR)))]
    return ls
```

```python
## Removing the outliers present in Price and Mileage Columns
index_list = []
for feature in ['price','mileage']:
    index_list.extend(remove_outlier_IQR(df, feature))
```

```python
## Finding the number of outliers
len(index_list)
```

```python
def remove(df, ls):
    ls=sorted(set(ls))
    df=df.drop(ls)
    return df
```

```python
## Dropping the outliers
df = remove(df,index_list)
```

```python
# Sepearating dependent and independent columns
y = df[['price']]
x = df.drop('price', axis=1)
```

```python
# Train-Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=25)
```

```python
x_train.shape
```

```python
# Modelling
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import r2_score
```

```python
models = {'Linear Regression' : LinearRegression(), 
          'Decision Tree' : DecisionTreeRegressor(), 
          'Random Forest' : RandomForestRegressor(),  
          'KNN' : KNeighborsRegressor()}
```

```python
for name, model in models.items() :
    print(name)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    
    print('Train Score', r2_score(y_train, model.predict(x_train)))
    print('Validation Score', r2_score(y_test, y_pred))
    print('\n')
```


#### Random Forest Regressor gives the best result among the other Regressor Models
